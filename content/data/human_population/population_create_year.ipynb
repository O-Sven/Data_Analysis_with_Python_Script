```python
for i in range(1960,2017):
    data = open("population.csv", "r")
    year_counter = str(i)
    data_new = open("human_population_{}.log".format(year_counter), "w+")
    for line in data:
        line_data = line.strip()
        line_data = str(line_data).split(",")
        country = line_data[0]
        year = line_data[2]
        population = line_data[3]

        if year == year_counter:
            line_data_new = country + "," + year + "," + population + "\n"
            print(line_data_new)
            data_new.write(line_data_new)

    data_new.close()
    data.close()
    
    
```
