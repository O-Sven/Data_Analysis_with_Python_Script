```python
def get_ball_position(filename):
    image = plt.imread(filename)#[:200, 1000:1200,:]
    r = image[:,:,0]
    g = image[:,:,1]
    b = image[:,:,2]

    r[r < 2*b] = 0
    r[r < 2*g] = 0
    r[r < 0.4] = 0

    x = 0
    y = 0
    n = 0
    for index, v in np.ndenumerate(r):
        if v > 0.01:
            y += index[0]
            x += index[1]
            n += 1
    
    return x/n, y/n
```

```python
import glob
filelist = sorted(glob.glob('data_red_ball/output_000[345678]??.png'))

x = []
y = []
for filename in filelist:
    print("processing file: ", filename)
    cx, cy = get_ball_position(filename)
    print(cx, cy)
    x.append(cx)
    y.append(cy)

print(x,y)
x = np.array(x)
y = np.array(y)
```

```python
back = plt.imread("./data_red_ball/output_000001.png")
plt.imshow(back)
plt.scatter(x,y)
```

```python
t = np.arange(300,883,1)
t
```

```python
plt.plot(t,x, label='x')
plt.plot(t,y, label='y')
plt.legend()
```

```python
vx_raw = np.gradient(x, t)
vx = np.convolve(vx_raw, np.ones(10)/10, mode='same')
vy_raw = np.gradient(y, t)
vy = np.convolve(vy_raw, np.ones(10)/10, mode='same')
v = np.sqrt(vx**2 + vy**2)
```

```python
plt.plot(t, v)
```

```python
s = np.cumsum(np.sqrt((x[1:]-x[:-1])**2 + (y[1:]-y[:-1])**2))
plt.plot(t[1:], s)
```

```python
ax_raw = np.gradient(vx, t)
ax = np.convolve(ax_raw, np.ones(10)/10, mode='same')
ay_raw = np.gradient(vy, t)
ay = np.convolve(ay_raw, np.ones(10)/10, mode='same')
a = np.sqrt(ax**2 + ay**2)
plt.plot(t,ax)
```

```python
from matplotlib.gridspec import GridSpec
file_index = 490
back = plt.imread("./data_red_ball/output_{:06d}.png".format(file_index))
index = np.where(t >= file_index)[0][0]

fig = plt.figure(figsize=(16,6))
gs = GridSpec(2,4)


plt.subplot(gs[0,0])
plt.plot(t[:index],v[:index], color='r')
plt.plot(t,v, color='r', alpha=0.4)
plt.axvline(x=t[index], color='k')
plt.xlabel('time /frame')
plt.ylabel('velocity /(pixel/frame)')

plt.subplot(gs[1,0])
plt.plot(t[:index],a[:index], color='g')
plt.plot(t,a, color='g', alpha=0.4)
plt.axvline(x=t[index], color='k')
plt.xlabel('time /frame')
plt.ylabel('acceleration /(pixel/frame)')

plt.subplot(gs[:,1:])
axis = plt.gca()
axis.imshow(back + (1-back)*0.7, cmap='Greys')

plt.scatter(x[:index],y[:index], color='b', s=1, alpha=0.7)

x_arrow_pos = x[index] + vx[index] / v[index]*40
y_arrow_pos = y[index] + vy[index] / v[index]*40
axis.arrow(x_arrow_pos, y_arrow_pos, 10*vx[index], 10*vy[index], color='r')
x_arrow_pos = x[index] + ax[index] / a[index]*40
y_arrow_pos = y[index] + ay[index] / a[index]*40
axis.arrow(x_arrow_pos, y_arrow_pos, 100*ax[index], 100*ay[index], color='g')

axis.add_artist(plt.Circle((x[index], y[index]), 40, color='b', fill=False))

plt.tight_layout()
```

```python
def annotate_write(file_index):
    back = plt.imread("./data_red_ball/output_{:06d}.png".format(file_index))
    index = np.where(t >= file_index)[0][0]
    
    t_start = 350
    
    fig = plt.figure(figsize=(12,6))
    gs = GridSpec(2,4)

    plt.subplot(gs[0,0])
    plt.plot(t[:index],v[:index], color='r')
    plt.plot(t,v, color='r', alpha=0.2)
    plt.axvline(x=t[index], color='k')
    plt.xlabel('time /frame')
    plt.ylabel('velocity /(pixel/frame)')

    plt.subplot(gs[1,0])
    plt.plot(t[:index],a[:index], color='g')
    plt.plot(t,a, color='g', alpha=0.2)
    plt.axvline(x=t[index], color='k')
    plt.xlabel('time /frame')
    plt.ylabel('acceleration /(pixel/frame^2)')

    plt.subplot(gs[:,1:])
    axis = plt.gca()
    blend = 0.7*(1+np.tanh((file_index-t_start+20)*0.2))/2. 
    print(blend)
    axis.imshow(back + (1-back)*blend, cmap='Greys')

    if file_index > t_start:
        plt.scatter(x[:index],y[:index], color='b', s=1, alpha=0.7)

        x_arrow_pos = x[index] + vx[index] / v[index]*40
        y_arrow_pos = y[index] + vy[index] / v[index]*40
        axis.arrow(x_arrow_pos, y_arrow_pos, 10*vx[index], 10*vy[index], color='r', width=3)
        x_arrow_pos = x[index] + ax[index] / a[index]*40
        y_arrow_pos = y[index] + ay[index] / a[index]*40
        axis.arrow(x_arrow_pos, y_arrow_pos, 100*ax[index], 100*ay[index], color='g', width=3)

        axis.add_artist(plt.Circle((x[index], y[index]), 40, color='b', fill=False))

    plt.tight_layout()
    
    plt.savefig("./out/output_{:06d}.png".format(file_index), dpi=300)
    plt.close()
```

```python
annotate_write(400)
```
